## 네이버에서 정보처리기사 검색 후 파워링크 부분 가져오기

```python
from bs4 import BeautifulSoup as bs
```

```python
#soup = bs(응답받은 데이터,해석프로그램)
#해석프로그램 : lxml(빠름), html.parser(중간), html5lib(느림)
url = 'https://search.naver.com/search.naver'
res = req.get(url,headers=headers,params={'query':'정보처리기사'})

soup = bs(res.text, 'html.parser')

#soup.find(태그명, 아이디 or 클래스) 
#네이버 파워링크 가져오기
# -> 요소(태그) 하나를 검색
powerLinkTag = soup.find('a',class_='lnk_tit') # 'a'만 있어도 돌아감. class옆에 _는 find 메소드의 서술 방식?임.

#텍스트 데이터만 가져오기
powerLinkKeyword = soup.find('a',class_='lnk_tit').text

# soup.find_all('a',class_='lnk_tit')
#len(리스트) : 리스트의 길이
#soup.findAll(): 여러 개의 요소를 검색한 후,
#                리스트로 반환 ( == soup.find_all() )
print(len(soup.find_all('a','lnk_tit')))
```

# 네이버 음악 Top50 수집(실습)

```python
url = 'https://music.naver.com/listen/top100.nhn?domain=TOTAL_V2'

res = req.get(url)

soup = bs(res.text,'html.parser')

#select(CSS선택자) : 여러 요소를 검색한 후 리스트로 반환
#                    --> find_all()과 같음

#select_one(CSS선택자) : 하나의 요소만 반환
#                    --> find()와 같음

rank_list = soup.find_all('td',class_='ranking')
name_list = soup.select('a._title > span')
# name2_list = soup.select('.ellipsis')
artist_list = soup.select('td.artist > a')
# artist2_list = soup.find_all('span',class_='ellipsis')
```

```python
len(rank_list), len(name_list), len(artist_list)
#len(artist2_list),len(name2_list)
```

```python
print(rank_list[0].text)
print(name_list[0].text)
print(artist_list[0].text.strip())
```

```python
ranks = []
names = []
artists = []
artists2 = []
for index in range(len(rank_list)):
    name = name_list[index].text
    rank = rank_list[index].text
    artist = artist_list[index].text.strip()
    
    names.append(name)
    ranks.append(rank)
    artists.append(artist)
```

```python
music_info = {'순위':ranks,'곡명':names,'아티스트':artists}
music = pd.DataFrame(music_info)
```

# 네이버 베스트셀러(실습) -- 과제

```python
url = 'https://book.naver.com/bestsell/bestseller_list.nhn'

res = req.get(url)
soup = bs(res.text,'lxml')
```

```python
titles = []
authors = []
grades = []
prices = []
dates = []
```

```python
titles_list = soup.findAll('a',class_='N=a:bel.title')
authors_list = soup.select('.type_best + dl > dt + dd')#저자가 ,로 구분되어있으면 밀리는현상이있으니 셀렉터로 가지고와서 함
grades_list = soup.select('.review_point > span')
prices_list = soup.findAll('em',class_='price')
dates_list = soup.select('.type_best + dl > dt + dd')
# dates_list = soup.select('#book_title_0 + dd')
# dates_list = soup.select('.thumb.type_best + dl > dt + dd')

print(titles_list[0].text)

authors_text = authors_list[0].text
idx = authors_text.index(" |")
print(authors_text[:idx].strip())

ra = float(grades_list[0]['style'][6:-2])/10
print(round(ra,2))

print(prices_list[0].text)

print(dates_list[0].text[-10:])
```

```python
len(titles_list), len(authors_list), len(grades_list), len(prices_list), len(dates_list)
```

```python
print('<네이버 베스트 셀러>')
for index in range(len(titles_list)):
    print(titles_list[index].text)

    authors_text = authors_list[index].text
    idx = authors_text.index(" |")
    print(authors_text[:idx].strip())
    
    ra = float(grades_list[index]['style'][6:-2])/10
    print(round(ra,2))
    
    print(prices_list[index].text)
    
    print(dates_list[index].text[-10:])
    print('-'*20)
    
    titles.append(authors_text[:idx].strip())
    authors.append(authors_text[:idx].strip())
    grades.append(round(ra,2))
    prices.append(prices_list[index].text)
    dates.append(dates_list[index].text[-10:])   
```

```python
book_info = {'title':titles,'author':authors,
             'grade':grades,'price':prices,
             'date':dates}
```

```python
import pandas as pd
```

```python
book_df = pd.DataFrame(book_info)
book_df.head()
```
