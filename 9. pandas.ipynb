# 데이터 분석 및 조작을 위한 라이브러리

```python
import pandas as pd
```

### 1차원 데이터와 2차원 데이터로 나누어져 있음
- 1차원 데이터(Series) : 인덱스와 값
- 2차원 데이터(Data Frame) : 표와 같은 데이터

```python
# Series 사용법 - 리스트 형태로 데이터 삽입
# 인덱스 지정 시엔 index = []로 삽입

population = pd.Series([9904312, 3448787, 2890451, 2466052],
                       index = ['서울', '부산', '인천', '대구'])
population
```

```python
# Series의 값 확인 --> .values
population.values

# Series 인덱스 확인 --> .index
print(population.index)

# Series의 타입 확인 --> dtype
print(population.dtype)
```

```python
# Series에 이름 지정
population.name = "인구"
population.index.name = "도시"

population
```

```python
# Series의 연산
print(population/100)
print(population*2)
```

```python
# Series의 인덱싱
print(population[0])
print(population[-1])
print(population[-2])
#==============================
# 인덱스 명으로 인덱싱
print(population["부산"])
```

```python
# 여러개 인덱싱
print(population[[0,3,1]])
print(population[['서울','인천']])
```

```python
# Series의 boolean 인덱싱

population > 2500000
```

```python
population[(population > 2000000) & (population < 5000000)]
```

```python
# Series의 슬라이싱
print(population[1:3])
print(population[1:])
print(population[ :3])
print(population[:])
```

```python
# 인덱스 명으로 슬라이싱
print(population["부산" : "대구"])
print(population[:"인천"])
```

```python
data_dic = {"A" : [1,3,3,4,4], "B" : [1,2,2,3,3], "C" : [1,2,4,4,5]}

df3 = pd.DataFrame(data_dic)
df3
```

```python
# apply함수 -- 행렬에 관련된 함수에 집어넣고 싶을 때
# NaN = 결측치
df4 = df3.apply(pd.value_counts)
df4
```

```python
# 결측치 채워주기
df4.fillna(value = 0)
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
df1
```

```python
df2
```

```python
df3
```

```python
#데이터 병합 - concat()
result = pd.concat([df1,df2,df3])
result
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'], 
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
df4
```

```python
# 아래로 합칠 때는 concat쓴다.
result = pd.concat([df1,df4])
result
```

```python
result_axis = pd.concat([df1,df4], axis = 1)
result_axis
```

```python
result_join = pd.concat([df1, df4], axis = 1, join = "inner")
# 결측치가 없는 인덱스만 병합했음.
result_join
```

```python
# ignore_index는 인덱스 번호를 완전 초기화 해서 다시 0부터 지정
result_ignore = pd.concat([df1,df4], ignore_index=True)
result_ignore
```

```python
df5 = pd.DataFrame({'key':['K0','K2','K3','K4'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','A3']})


df6 = pd.DataFrame({'key':['K0','K1','K2','K3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})
df5
```

```python
df6
```

```python
#옆으로 합칠 때는 merge쓴다.
result_merge = pd.merge(df5, df6)
result_merge
```

```python
# 오른쪽 , 즉 df6를 기준으로 합병, 따라서 df6가 가지고 있지 않은 데이터가 Nan값으로 뜸
re_right = pd.merge(df5, df6, how = "right")
re_right
```

```python
# merge보다는 concat을 더 많이 쓴다.
re_left = pd.merge(df5, df6, how = "left")
re_left
```
