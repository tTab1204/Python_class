# Dictionary
- 기본 구조 --> {key:value}

```python
print(type(dic_test))
```

```python
#딕셔너리의 key값은 중복되어서는 안되며, 오직 한개만 있어야 함
#존재하는 dictionary에 값 추가
#딕셔너리명[key] = value값
dic_test = {"이름" : "원호", "나이" : 20, "전화번호" : "010-0000-1111"}
dic_test["생일"] = "1월 28일"
dic_test
```

```python
# 존재하는 딕셔너리에 값 삭제
# del 딕셔너리명[key]
dic_test = {"이름" : "원호", "나이" : 20, "전화번호" : "010-0000-1111", "직업" : "강사"}
del dic_test["나이"]
dic_test
```

```python
#값 가져오기 - 1번
#딕셔너리명 [key]
dic_test["이름"]
```

```python
#값 가져오기 -- 2번
#딕셔너리명.get(key)
dic_test.get("이름")
```

```python
#1번과 2번의 차이점
dic_test["혈액형"] #없는 key값 입력시 에러가 남
```

```python
a = dic_test.get("혈액형") #없는 key값 입력 시 false
print(a)
```

```python
#딕셔너리에 존재하는 key값만 가져오기
type(dic_test.keys())

list(dic_test.keys())
```

```python
#딕셔너리에 존재하는 value값만 가져오기
type(dic_test.values()) #dict_values

list(dic_test.values()) # dict_values 타입을 list타입으로 바꿈
```

```python
# for문 활용하여 요소 뽑기
for key in dic_test.keys():
    print(key)
```

```python
for value in dic_test.values():
    print(value)
```

```python
dic2 = {"이름":[1,2,3,4]}
dic2["이름"].append(5)
dic2
```

```python
#for문을 활용해서 동시에 key값과 value값을 가져오기
for k, v in dic_test.items():
    print(k,v)
```

```python
# in 키워드
# key in 딕셔너리명 --> True or False
# Key값만 가능

"혈액형" in dic_test
```
