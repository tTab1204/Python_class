```python
# 파일 읽기 -- readline() 반복문 사용해서 읽기
f = open("C:/Users/vdi02/Desktop/Python_file/새파일.txt","r")

while True:
    line = f.readline()
    if not line: #값이 없을 시 if문이 True
        break
    print(line)
f.close()
```

```python
# 파일 읽기 -- 2. readlines()함수 사용
f = open("C:/Users/vdi02/Desktop/Python_file/새파일.txt","r")
lines = f.readlines()
print(lines)

f.close()
```

```python
# 파일 읽기 -- 2.1 readlines()함수를 반복문 사용하여 한줄 씩 읽어오기, 리스트 타입으로 저장
f = open("C:/Users/vdi02/Desktop/Python_file/새파일.txt","r")
lines = f.readlines()
for i in lines:
    print(i)
f.close()
```
