# 반복문
- 특정조건에 따라 실행문장을 반복해서 실행하는 제어구문
- 종류: while, for-each, range()함수를 활용한 for문

```python
list_food = ["햄버거", "치킨", "피자"]

#for-each문
# for 변수명 in 리스트:
#     실행문장

for food in list_food:
    print(food)
```

```python
score_list = [90, 45, 70, 60, 55]
number = 0
for score in score_list:
    number+=1
    if score >= 60:
       print("{}점, {}번 학생은 합격입니다.".format(score, number))
    else:
        print("{}점, {}번 학생은 불합격입니다.".format(score,number))
```

```python
# 1 ~ 10까지 출력하시오.
# for i in range(1, 11, 1): # i <= i < 11
#     print(i)
#증가값 위치는 생략이 가능
# for i in range (1, 11):
#     print(i)

# 시작값 위치 생략 가능
# print()함수의 end속성: 기본값 \n(줄바꿈), 새로운 값을 지정하면 줄바꿈이 사라짐.
# shift + tab --> 함수 설명 띄우기

for i in range(10):
    print(i+1, end=" ")
```
