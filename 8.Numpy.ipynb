```python
with open("height_weight.txt","r") as f:
    height = f.readline()
    weight = f.readline()
    
height = height.split(",")
weight = weight.split(",")
print(height)
print(weight)
```

```python
#dtype np.array의 타입을 바꿔주는 속성
arrayHeight = np.array(height, dtype = np.float)
arrayWeight = np.array(weight, dtype = np.float)

arrayHeight /= 100

BMI = arrayWeight / (arrayHeight**2)
print(BMI)
```

```python
#인덱싱
arr1 = np.arange(10)
print(arr1[0])
print(arr1[-1])
print(arr1[-2])
```

```python
# 슬라이싱
print(arr1[2:5])
print(arr1[2:])
print(arr1[:5])
print(arr1[:]) #전체 출력하기

#인덱싱, 슬라이싱 --> 문자열, 리스트 모두 동일
```

```python
arr2 = np.arange(50).reshape(5,10) #reshape(행, 열)
arr2
```

```python
print(arr2[2:4]) # 2~4 번째 미만의 행
print("===================")
print(arr2[2:]) # 2번째 행부터 ~끝까지
print("===================")
print(arr2[:4]) # 처음부터 ~ 4번째 미만까지
print("===================")
print(arr2[:]) # 전체출력 
```

```python
print(arr2[:,:]) # ,(쉼표)를 기준으로 행/열로 나누어진다.
print("===================")
print(arr2[:2,:]) # 2번째 행 미만의 값을 모두 출력
print("===================")
print(arr2[:,0]) # 모든 행의 0번째 열만 출력
print("===================")
print(arr2[:4, :5]) # 4행 5열 안에 있는 모든 수
print("===================")
print(arr2[2,1]) # 2행 1열에 있는 값을 가져와라
print("===================")
print(arr2[[2,2],[3,4]]) #여러개의 값 가져오는 것도 가능
```
